```python
import pandas as pd
from collections import Counter
import re
```

```python
# to import files from googledrive
from google.colab import drive


drive.mount('/content/gdrive/', force_remount=True)
sample_size_dir = 'gdrive/My Drive/Sample size for NLP'
```

```python
diagnosis = '5849'
```

```python
df = pd.read_csv(sample_size_dir+'/data_jaya/'+diagnosis+'_combined_data_1000.csv')
# df = pd.read_csv(sample_size_dir+'/Sample_Size_Saniya/12k_data_for_classifier.csv')
df.head()
```

```python
# Function to tokenize text
def tokenize(text):
    tokens = re.findall(r'\b\w+\b', text.lower())
    return tokens
```

```python
# Tokenize texts for each label
df['tokens'] = df['TEXT'].apply(tokenize)
# df['tokens'] = df['text'].apply(tokenize)
```

```python
# Separate rows by label
df_label_0 = df[df['label'] == 0]
df_label_1 = df[df['label'] == 1]
```

```python
# Build sets of unique tokens for each label
vocabulary_label_0 = set([token for tokens in df_label_0['tokens'] for token in tokens])
vocabulary_label_1 = set([token for tokens in df_label_1['tokens'] for token in tokens])
```

```python
# Calculate vocabulary sizes
vocabulary_size_label_0 = len(vocabulary_label_0)
vocabulary_size_label_1 = len(vocabulary_label_1)
```

```python
print(f"Vocabulary Size for Label 0: {vocabulary_size_label_0}")
print(f"Vocabulary Size for Label 1: {vocabulary_size_label_1}")
```

```python
# Save the results to a text file
with open(sample_size_dir+'/data_jaya/'+diagnosis+'_vocabulary_sizes.txt', 'w') as f:
    f.write(f"Vocabulary Size for Label 0: {vocabulary_size_label_0}\n")
    f.write(f"Vocabulary Size for Label 1: {vocabulary_size_label_1}\n")
```

```python
print(f"Vocabulary for Label 0: {vocabulary_label_0}")
```

```python
print(f"Vocabulary for Label 1: {vocabulary_label_1}")
```

```python
# Save the results to a text file
with open(sample_size_dir+'/data_jaya/'+diagnosis+'_vocabulary_sizes_and_words.txt', 'w') as f:
    f.write(f"Vocabulary Size for Label 0: {vocabulary_size_label_0}\n")
    f.write("Vocabulary for Label 0:\n")
    f.write(", ".join(sorted(vocabulary_label_0)) + "\n\n")

    f.write(f"Vocabulary Size for Label 1: {vocabulary_size_label_1}\n")
    f.write("Vocabulary for Label 1:\n")
    f.write(", ".join(sorted(vocabulary_label_1)) + "\n")
```

```python
vocab_list_0 = sorted(list(vocabulary_label_0))
vocab_list_1 = sorted(list(vocabulary_label_1))
```

```python
# Create a DataFrame with the vocabularies
max_length = max(len(vocab_list_0), len(vocab_list_1))
vocab_list_0.extend([''] * (max_length - len(vocab_list_0)))  # Pad with empty strings
vocab_list_1.extend([''] * (max_length - len(vocab_list_1)))  # Pad with empty strings

vocab_df = pd.DataFrame({
    'Label 0 Vocabulary': vocab_list_0,
    'Label 1 Vocabulary': vocab_list_1
})

vocab_df.head()
```

```python
vocab_df.to_csv(sample_size_dir+'/data_jaya/'+diagnosis+'_vocabularies.csv')
```
